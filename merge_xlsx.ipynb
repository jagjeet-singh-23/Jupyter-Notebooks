```python
import pandas as pd
```

```python
xlsx = pd.ExcelFile('/content/401R_fake_data.xlsx')
```

```python
df1 = pd.read_excel(xlsx, '401R')
df1.dropna(inplace=True)
df1.head()
```

```python
df2 = pd.read_excel(xlsx, 'JOB CARD REPORT')
df2.dropna(inplace=True)
df2
```
